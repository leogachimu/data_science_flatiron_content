## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
q = '''SELECT o.officeCode, city, employeeNumber
        FROM offices AS o
        LEFT JOIN employees AS e
        ON o.officeCode = e.officeCode
        WHERE employeeNumber IS NULL
        GROUP BY o.officeCode
'''
pd.read_sql(q, conn)
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

```python
# Firstly, link employees to customers
# secondly, link customers to orders
# thirdly, link orders to order details
# lastly, link order details to products
q_productNames = '''SELECT employeeNumber, firstName, lastName, productName
        FROM employees AS e 
        JOIN customers AS c
        ON e.employeeNumber = c.salesRepEmployeeNumber
        JOIN orders 
        USING(customerNumber)
        JOIN orderdetails as od
        USING(orderNumber)
        JOIN products as p
        USING(productCode)
       
'''
product_per_employee_df = pd.read_sql(q_productNames, conn)
product_per_employee_df
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
# Firstly, link employees to customers
# secondly, link customers to orders
# lastly, link orders to order details
q_orderQuantity = '''SELECT employeeNumber, firstName, lastName, SUM(quantityOrdered) AS num_of_products
        FROM employees AS e 
        JOIN customers AS c
        ON e.employeeNumber = c.salesRepEmployeeNumber
        JOIN orders 
        USING(customerNumber)
        JOIN orderdetails as od
        USING(orderNumber)
        GROUP BY e.employeeNumber
        ORDER BY lastName
'''
pd.read_sql(q_orderQuantity, conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Firstly, link employees to customers
# secondly, link customers to orders
# thirdly, link orders to order details
# lastly, link order details to products
q_productNames = '''SELECT employeeNumber, firstName, lastName, COUNT(productName)
        FROM employees AS e 
        JOIN customers AS c
        ON e.employeeNumber = c.salesRepEmployeeNumber
        JOIN orders 
        USING(customerNumber)
        JOIN orderdetails as od
        USING(orderNumber)
        JOIN products as p
        USING(productCode)
        GROUP BY e.employeeNumber
        HAVING COUNT(productName) > 200
'''
pd.read_sql(q_productNames, conn)
```
